```python
import pandas as pd
import cleaning



df1 = pd.read_csv('sneakers_adidas.csv')
df2 = pd.read_csv('sneakers_jordan.csv')
df3 = pd.read_csv('sneakers_nike.csv')
df4 = pd.read_csv('sneakers_puma.csv')
df5 = pd.read_csv('sneakers_mizuno.csv')
```

```python
api_df = pd.concat([df1, df2, df3, df4], ignore_index = True)

api_df.shape
```

```python
df5.shape
```

```python
api_df.head()
```

```python
columns_to_drop = ['id', 'image', 'links', 'releaseDate', 'releaseYear', 'sku', 'story']
api_df = cleaning.drop_columns(api_df, columns_to_drop)
api_df
```

```python
api_df = api_df.rename(columns = {'colorway' : 'primary_color'})
api_df['primary_color'] = api_df['primary_color'].str.split('/').str[0]
api_df['primary_color'].nunique()
```

```python
df5 = df5[df5['primary_color'] != 'Lightning Star Jr']
df5 = df5[df5['primary_color'] != 'Wos']
df5['primary_color'].unique()
```

```python
color_changes = {'Preto':'Black',
                 'Preta':'Black',
                 'Branca':'White',
                 'Branco':'White',
                 'Azul':'Blue',
                 'Amarelo':'Yellow',
                 'Roxo':'Purple',
                 'Br':'White',
                 'Rosa':'Pink',
                 'Bo':'Burgundy',
                 'Laranja':'Orange',
                 'Cinza':'Grey'
                 }
df5['primary_color'] = df5['primary_color'].replace(color_changes)
df5['primary_color'].unique()

df5
```

```python
df = pd.concat([api_df, df5], ignore_index = True)
df
```

```python
cleaning.standardize_column_names(df)
df
```

```python
cleaning.unique_values_for_categorical(df, 30)
```

```python
df = df[df['gender'] != 'Another_One']
```

```python
replacement_dict = {'child': 'kid', 'youth': 'kid', 'infant': 'kid', 'toddler': 'kid', 'preschool': 'kid'}
df['gender'] = df['gender'].replace(replacement_dict)
df['gender'].unique()
```

```python
df.to_csv('sneakers.csv', index=False)
```
